```python
#FOR COUNTRIES FROM STARTING OF THE DATA
# Create the bar plot                       #YEAR 1985
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="1985", y="Country Code", data=df.head(10), palette="coolwarm")
plt.title("Top 5 Countries of Population Data(1985)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot                      #YEAR 2023
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2023", y="Country Code", data=df.head(10), palette="coolwarm")
plt.title("Top 5 Countries of Population Data(2023)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
#FOR COUNTRIES FROM THE BOTTOM OF THE DATA
# Create the bar plot                   #YEAR 1985
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="1985", y="Country Code", data=df.tail(10), palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (1985)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot                   #YEAR 2023
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2023", y="Country Code", data=df.tail(10), palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2023)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
